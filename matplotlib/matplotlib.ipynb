```python
import pandas as pd
```

```python
data = pd.read_csv("Standard Metropolitan Areas Data.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data.crime_rate, data.percent_senior)
plt.title("Plot of Crime Rate vs Percent Senior")
plt.xlabel("Crime Rate")
plt.ylabel("Percent Senior")
```

```python
plt.plot(data.work_force, data.income)
plt.xlabel("Work Force")
plt.ylabel("Income")
```

```python
plt.figure(figsize=(12,5))
plt.plot(data.work_force, data.income)
plt.xlabel("Work Force")
plt.ylabel("Income")
```

```python
plt.plot(data.work_force, data.income, linestyle='--', marker='o', color='r')
plt.xlabel("Work Force")
plt.ylabel("Income")
```

```python
plt.plot(data.work_force, data.income, '--ro')
plt.xlabel("Work Force")
plt.ylabel("Income")
```

```python
plt.plot(data.work_force, data.income, 'r', label='Labor')
plt.plot(data.physicians, data.income, 'g', label='Graduates')
plt.legend()
```

```python
plt.subplot(1,2,1)
plt.plot(data.work_force, data.income, 'go')
plt.title("Income vs Work Force")

plt.subplot(1,2,2).label_outer() # use first plot labels
plt.plot(data.hospital_beds, data.income, "r^")
plt.title("Income vs Hospital Beds")

plt.suptitle("Sub Plots")
```

```python
fig, ax = plt.subplots(1,2,figsize=(6,6))
ax[0].plot(data.work_force, data.income, 'go')
ax[1].plot(data.hospital_beds, data.income, "r^")
```

```python
plt.title("Histogram")
plt.xlabel("Percentage of Senior Citizens")
plt.ylabel("Frequency")

plt.hist(data.percent_senior)
```
